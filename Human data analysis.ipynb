````python
import pandas as pd
file_path = 'data/campaign.csv'
df = pd.read_csv(file_path)
df.head()
````

````python
df.describe(include='all')
````

### Questions

- Lots of NULLs in ```'previous_outcome'```, why?

- I'd guess ```'previous_outcome'``` is a past success or failure with that customer 

    - Also that a null in outcome would correlate with a 0 ```'previous_campaign_contacts'``` 

    - Accurate guess?



````python
# what is the range of values that both columns have

unique_pcc = pd.unique(df['previous_campaign_contacts'])
unique_pcc
# array([0, 1, 2, 3, 4, 5, 6, 7], dtype=int64)

unique_prev_outcome = pd.unique(df['previous_outcome'])
unique_prev_outcome
# array([nan,  0.,  1.])

# Find unique values of outcome when previous_campaign_contacts is 0
unique_prev_outcome_if_pcc_0 = df[df['previous_campaign_contacts'] == 0]['previous_outcome'].unique()
unique_prev_outcome_if_pcc_0

# array([nan])

# Find uniq vals of outcome if pcc is greater than 0
unique_prev_outcome_if_pcc_1 = df[df['previous_campaign_contacts'] > 0]['previous_outcome'].unique()
unique_prev_outcome_if_pcc_1

# array([0., 1.])
````
